```python
import pandas as pd
"""
問題一
讀入 excel 試算表：https://storage.googleapis.com/py_ml_datasets/Energy%20Indicators.xls 並儲存成 DataFrame，命名為 energy
擷取這四個變數並重新命名：['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
將 Energy Supply 單位改為 gigajoules(1 petajoule = 1,000,000 gigajoules)
將 ... 的資料記錄為遺漏值 np.NaN
重新命名國家名稱：
"Republic of Korea": "South Korea"
"United States of America": "United States"
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom"
"China, Hong Kong Special Administrative Region": "Hong Kong"
移除小括號或數字，像是 'Bolivia (Plurinational State of)': 'Bolivia'、'Switzerland17': 'Switzerland'
"""

def hasNumbers(inputString):
  return any(char.isdigit() for char in inputString)


def remove_brackets_content(str):
    new_s = ""
    
    for char in str :
          if(char !='('   ):
            new_s+=char
          else:
            break
    return new_s        
        


df = pd.read_excel("Energy Indicators.xls")
#print(len(energy.index.values))

for i in range(len(df.index.values)):
    
    country_name = df.iloc[i,2]
    #print("i=",i)
    #print("before =", country_name)
    if (country_name != country_name):
       country_name = "np.NaN"   
    elif (hasNumbers(country_name) == True): 
      
        country_name = ''.join([i for i in country_name if not i.isdigit()])
        #print("with number =", country_name)
    elif (country_name.find('(')!=-1 and  country_name.find(')')!=-1):
        country_name = remove_brackets_content(country_name) 
        #print("with () =", country_name)
    else:
        pass
    
    #print("after",country_name)
    
    if (country_name == "np.NaN"):
        pass
       #print("country_name = ",country_name)
    elif (country_name == "Republic of Korea"):
        country_name = "South Korea"
        #print("country_name = ",country_name)
    elif (country_name == "United States of America"):
        country_name = "United States"
        #print("country_name = ",country_name)
    elif (country_name == "United Kingdom of Great Britain and Northern Ireland"): 
        country_name = "United Kingdom" 
        #print("country_name = ",country_name)
    elif (country_name == "China, Hong Kong Special Administrative Region"): 
        country_name = "Hong Kong"        
        #print("country_name = ",country_name)
    else:    
        pass
    
    df.iloc[i,2] = country_name
    #print(country_name)
    #print(df.iloc[16:,2:])
    

energy = pd.DataFrame()
energy['Country'] = df.iloc[16:243,2]
energy['Energy Supply'] = df.iloc[16:243,3]
energy['Energy Supply per Capita'] = df.iloc[16:243,4]
energy['% Renewable'] = df.iloc[16:243,5]

print(energy)
```

問題十三
將人口估算變數改為字串，並且利用逗號分隔千分位數。

e.g. 317615384.61538464 -> 317,615,384.61538464

```python
def get_comma_format(x):
    '''
    將數字轉換為千分位格式的文字
    '''
    return '{:,}'.format(x)
############################################
#問題十三
#將人口估算變數改為字串，並且利用逗號分隔千分位數。
#
#e.g. 317615384.61538464 -> 317,615,384.61538464
############################################
def answer_thirteen():
    '''
    這個函數應該回傳一個 Series PopEst，索引值為國家名稱
    '''
    Top15 = answer_one()
    Top15['PopEst'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Pop_list = []
    for i in range(len(Top15['PopEst'])):
       Pop_list.append(get_comma_format(Top15['PopEst'][i]))
        
    Top15['PopEst'] = Pop_list    
    #print(Top15['Pop'])

    return Top15['PopEst']
answer_thirteen()
```
